# シグモイド関数とは

シグモイド関数（別名：ロジスティック関数） g(z)は以下の様な関数です。

$g(z) = \dfrac{1}{1 + e^{-z}}$


```python

import math

x = np.arange(-10, 10, 0.1)
e = math.e

y = 1 / (1 + e**-x)
fig = plt.figure(figsize=(12,8))
plt.plot(x, y)
plt.xlabel("dot(theta.T, X)")
plt.ylabel("probability")
plt.title("Sigmoid function")
plt.show()
```

## 関数定義する

```python
#シグモイドを関数にしてみる
#exp:指数関数 のこと
def sigmoid(z):
    s = 1 / (1 + np.exp(-z))
    return s

```

# ロジスティクス回帰（コスト計算準備含む）

ロジスティック回帰における仮説$h_\theta (x)$は、シグモイド関数$g(x)$に予測変数 x
 のパラメーター$z = \theta^T x$を代入します。

${ h_\theta (x)  = \frac{1}{1 + e^{-\theta^{\mathrm{T}} x}} } $

## 交差エントロピー誤差とは

交差エントロピーとは、2つの確率分布の間に定義される尺度である。符号化方式が、真の確率分布ではなく、ある所定の確率分布に基づいている場合に、とりうる複数の事象の中からひとつの事象を特定するために必要となる数の平均値を表しています。

```python
def compute_cost(X, y, theta, lamb):
    #内積を計算
    dot = np.dot(X,theta)
    #シグモイドに内積の値を入れます
    h = sigmoid(dot)
    # 行の数をとってくる
    m = len(X)
    # Log0を防ぐための策
    prev = pd.DataFrame([0,1,1,1,1])
    #途中でDateFrameを使って無理やり計算
    J = (1/m) * ( np.dot(-y.T , np.log(h)) - np.dot((1 - y).T , np.log(1-h)) ) + (lambda1/(2*m)) * ((theta*theta)*prev).sum().values
    return J[0,0]
```

```python
#計算
compute_cost(x_df, y_df, theta, 0.01)
```

# 最急降下法(勾配降下法、GradientDescent)

```python
def gradient_descent(X, y, theta, iterations, alpha):
    #ここにいれていく
    m = len(X)
    past_costs =[]
    past_thetas = [theta]
    
    for i in range(iterations):
        dot = np.dot(X, theta)
        #hとxから
        h_y = sigmoid(dot) - np.array(y)
        
        gradient = np.dot(X.T, h_y)

        #thetaを更新
        theta = theta - (alpha/m) * (gradient + lambda1* theta * pd.DataFrame([0,1,1,1,1]))
        # New Cost
        new_cost = compute_cost(X, y, theta, lamb)
        
        # thetaとcostを保持する変数past_costsとpast_thetasに順次過去の値をappendする
        past_costs.append(new_cost)
        past_thetas.append(theta)    
        
    return past_costs, past_thetas
```

```python
# 確率を求める場合の関数
def predict_probs(X, theta):
    theta_x = np.dot(X,theta)
    return sigmoid(theta_x)
    
    
# 分類を行う場合の関数
def predict(X, theta, threshold=0.5):
    theta_x = np.dot(X,theta)
    pred=(sigmoid(theta_x)>=threshold)
    return pred
```

# なぜシグモイドを用いるか

ロジスティック回帰では予測する出力が数値ではなくカテゴリーに属するかどうかの確率になります。

そのため下記３つの条件をみたすシグモイド関数がよく用いられます。
- $h_\theta(x)$がどんな値を取ったときも0から1の値を取ることができる。
- 単調増加（$x1 < x2$のとき$h_\theta(x1) < h_\theta(x2)$が成り立つ。）のため値の大小関係は崩れない。
- $h_\theta(x)$の正負で分類が可能であり、判定が楽。

# 正則化：「あなたに似た人を見つけることができました。それはあなたです」を防ぐ技術

機械学習では大量のデータを扱いますが、偏り過ぎたデータにまで必要以上に対応してしまうという「過学習」の状態に陥る事があります。

過学習とは、「あなたに似た人を見つけることができました。それはあなたです」という状態だと言われます。

与えた学習データに対しては小さな誤差となるモデルが構築できています。ただし、ごく一部の例外的な学習データに過度に対応したモデルとなっているために、構築した学習モデルを未知データに適用すると必ずしも適切な予測値を返さない状態となります。学習データの中のごく一部の例外的なデータに過度に適用したモデルが構築されている状態ということで「過学習」というのですが、

一般的に使用されるのは、L1正則化とL2正則化のようです。
学習モデルのパラメータのL1ノルム、L2ノルムを用いるものです。別の言い方をすると、L1正則化は、ペナルティとして学習モデルのパラメータの絶対値の総和を用いるものであり、L2正則化は、ペナルティとして学習モデルのパラメータの二乗の総和を用いるもので、それぞれ下記の特徴をもちます。

 - L1正則化　特定のデータの重みを0にする事で、不要なデータを削除する
 - L2正則化　データの大きさに応じて0に近づけて、滑らかなモデルとする

つまり、L1正則化もL2正則化も、モデルの次元が低い方がペナルティが小さくなる正則化のモデルです。従って、学習データの中に例外的なデータが含まれていた場合に、その例外的なデータに対応するために学習モデルの次元をいくつも増やす必要がある場合には、その例外データには対応しないようなモデルが選択されます。

# ロジスティック回帰が交差エントロピー誤差を使う理由

ロジスティクス回帰は、分類を扱い、また交差エントロピー誤差は、正解の選択肢に対してのみ、出力と理想的な値との距離を測ることができるため。0と1での正解不正解の判定の誤差に使いやすいことが挙げられます。
計算式には全ての値が含まれますが、正解ラベルがone-hot表現されていれば不正解となる選択肢の項は自動的に全て消える仕様です。


例えば出力 y はクラスが5つの場合次のような列になります。
[0.1, 0.3, 0.8, 0.2, 0,1]　　 

正解のクラスが、1つ目は3(0.8)で2つ目は2(0.3)であるとき、教師データはそれぞれ次のようになります。
one-hot [0, 0, 1, 0, 0] 

出力 y と one-hot表現を掛けた場合
[log(0.1), log(0.3), log(0.8), log(0.2), log(0,1)] と one-hot [0, 0, 1, 0, 0] 
3つ目の値以外は 0 を掛けるために 0 になってしまい、log (0.8) だけが出てきて、計算結果は -log(0.8) になります。
