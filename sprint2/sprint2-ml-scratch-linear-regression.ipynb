# 損失関数定義：cost_function

```python
def compute_cost(X, y, theta):

    #sample 数
    ids = len(X)
    #naiseki
    naiseki = np.dot(X, theta)
    #distance
    distance = (naiseki - y) # (Hypothesis value -- Actual value)
    #distance**2 の合計
    total_distance = (distance**2).sum()
    #total_distanceの平均
    J = total_distance / (2 * ids)
    
    return J
```

# 最急降下法(勾配降下法、GradientDescent)

```python
def gradient_descent(X, y, theta, iterations, alpha):

    #ここにいれていく
    past_costs = []
    past_thetas = [theta]
    m = len(X)
    
    for i in range(iterations):
        #距離を出す
        distance = (np.dot(X, theta) - y) 
        #勾配を出す
        gradient = np.dot(X.T, distance) /m
        #thetaを更新
        theta = theta - (alpha * gradient)
        # New Cost
        new_cost = compute_cost(X, y, theta)

        # thetaとcostを保持する変数past_costsとpast_thetasに順次過去の値をappendする
        past_costs.append(new_cost)
        past_thetas.append(theta)
    return past_costs, past_thetas
```

### 教師あり学習：
機械はどうやって学習するのでしょうか？ 代表的な学習の仕方が、機械に質問と回答を同時に教える方法です。例えば、犬の画像に「分類は犬である」という正解を付けたデータを分析させます。これを「ラベル付きデータ」（正解付きのデータ）と呼びます。
では、正解がわかっているのに機械は何を分析するのかというと「どうしてそれが犬に分類されるのか？」ということを考えさせます。つまり機械に「特徴量」を自ら発見させる→それは「犬」に分類される→その特徴量のものは「犬」である、ということです。訓練データとして、正解付きの画像（ラベル付きデータ）を入力する。正解は犬だとわかっているので、機械は分析した特徴量が「犬」の特徴であることを学習し、大量に学習することで犬に分類すべき特徴量が蓄積され、精度が上がっていく。
しかし、画像を一枚だけ解析して特徴量を覚えたシステムだと、ほかの犬の画像を見せても、よほど似た犬の画像でなければ同じ犬の種類だとは判別できないでしょう。では、犬のラベル（正解）付きデータ画像を、仮に1000万枚読み込ませて訓練させたとしたらどうでしょうか。1000万通りの犬の特徴量を学習すれば、膨大な特徴量から照合することで、「犬」を識別できるようになる確率は高くなるでしょう。
学習するためには膨大なビッグデータが必要になる理由です。この犬の例のように、ラベル付きのデータで学習させる方法を「教師あり学習」と呼びます。


### 分類問題：
上記に書いた機械による学習では犬の教師あり学習について説明しました。このように犬であるか、犬でないか、またはそれを応用して、犬や他の動物等を区別していく問題を分類問題と呼びます。
分類問題は、機械学習以前から当然存在し、ロジスティック回帰などの統計学技術によって、複雑に入り組んだ現象を上手に切り分けてきました。


### 回帰問題：
過去の売り上げデータなどから未来の売り上げを予測するような数字を扱った問題を回帰問題と言います。

### 教師なし学習：
入力画像はあるけれど、その正解データは与えられないものです。予測は未来のことなので正解はありません。このように推論、分析など、正解がない、正解が解らない問題で学習することを教師なし学習と呼びます。


### 半教師あり学習：
教師あり学習と教師なし学習を混在させる方法です。まずは機械に対して教師あり学習で特徴量を学ばせて、それ以降は教師なし学習で膨大な訓練データを与え、自動的に特徴量を算出させながら繰り返し学習する方法です。これを「半教師あり学習」と呼ぶこともあります。


### 強化学習：
強化学習は、教師付き学習とよく似ているのですが、与えられた正解の出力をそのまま学習すれば良いわけではなく、もっと広い意味での「価値」を最大化する行動を学習すること目的としています。
例えば、テトリスでできるだけ高スコアを得るような問題も強化学習の枠組みで考えることができます。
その時点で一番スコアが高くなるのは、一列でもすぐに消すようなプレイ方法ですが、より長期的には、できるだけ溜めてから一度にたくさんの列を消したほうがスコアが高くなるというようなことができることもあります。

### 正規化：

例で説明します。

1,2,3,4,5という 5つのデータを、最大値が 1最小値が 0になるように変形すると、

$$ \left(0, 0.25, 0.5, 0.75, 1\right) $$

となります。



### 標準化：

1,2,3,4,5という 5つのデータを標準化してみましょう

$$ \left(\dfrac{-2}{\sqrt{2}},\dfrac{-1}{\sqrt{2}},0,\dfrac{1}{\sqrt{2}},\dfrac{2}{\sqrt{2}}\right) $$となります。不思議でしょう？
 

これは、$$ Y=\dfrac{X-\mu}{\sigma} $$という式で計算しているからです。

### 正規化と標準化の使い分け

正規化は、つまり割合で表示しており、複数の変数同士を比較しやすいです。一方標準化は、元のデータの分布を意識してみたいときに使い分けます。
ここも例で示しましょう。

標準化のメリットを具体例で説明します。5人でテストを受けたとしましょう。

同じ50点でも、周り4人が
10,20,30,40点の場合は「平均 +20点」の良い点数」ですが、
30,40,60,70点の場合は平均点ぴったりです。
つまり、前者の方が価値ある 50点です。
このように、複数の状況の数字を比較するときには、データの数字そのものではなく、平均を基準にして考える必要があります。

また、同じ 50点でも、周り4人が
10,20,30,40点の場合よりも
25,25,25,25点の場合の方が「珍しい、価値ある 50点」と考えることができます。

このように分布のどの位置にあるのかというのは、正規化ではわかりかねます。



### フィーチャースケーリングのメリット

ここでは、フィーチャースケーリングがないと逆にどのようなことになるのかを説明し、デメリットがあるため、使わない手はないという説明をします。
結論から言うと、特徴間でスケールがそろっていないと、学習モデルのパラメータ間で更新幅にかたよりが生じてしまうのです。

先ほどの正規化例では、外れ値がそこに入っていても、25％。あるいは0.25として扱われるので、それを使った学習では特徴量の分布を考慮できず、結果出てくる重み付けに正確な値を使えないことになります。


## 切片の意味
線形回帰モデルは、切片が存在する。 線形回帰モデルにおいてどのように役立っているか述べよ。

ヒントとして、切片が存在しない場合、モデルの表現力はどうなるかということが挙げられる。

たとえばある会社が、生産部門で働く従業員の業務遂行能力を
$$ y = 130 + 4.3x $$ 
という回帰モデルを使用して予測できると判断したとします。ここで、xは従業員が受ける社内教育の時間（0～20）を表し、yは職業技能検定での従業員のスコアを表します。y切片の値（130）は、社内教育を受けていない従業員の平均職業技能スコアを示しています。傾きの値（4.3）は、社内教育の時間が増えるごとに、職業技能スコアが平均で4.3ポイント増加することを示しています。切片が存在しない場合、平均的な職業技能スコアを無視すると言うことになり、傾きだけで目的変数を達成する必要があると言うことにもなる。それはモデルの表現として正確ではないし、そこから導き出されたものであるとよりきつい傾きを求められることになり、社員の職業技能がより早く獲得できるという誤った分析結果を提供してしまったりします。

## 学習率
どのような値から選択するべきか、それはなぜか

- Overfitting
- Underfitting

学習率は、率であるので０から１の間の割合の値を代入します。
大きい値を入れた時は、今いる位置から曲線の遠いところに一度に移動できるということを表し、次数の高い、または多項式の曲線上の場合に、早く求めたい値のあるところに行ける可能性も出てきます。しかし、それが求めたい最も低い値になるとは限りません。

そのため、最適な学習率はデータ（ネットワークのアーキテクチャ）によって異なるのですが、非常に小さい値( 最適なレンジは0.02～とされる )からスタートしていくことが多いということになっています。学習率が大きすぎたり小さすぎたりすると、ネットワークの学習状態が悪い(UnderFitting)、遅い、または全く学習しない、学習しすぎる(OverFitting)という結果を招きます。最適な桁数を発見したのち、有効数字2桁圏内で、最もパフォーマンスのよい数値を決めます。

学習曲線は、その名前の通り、機械学習のモデルが到達点までどのように推移しているか（学習しているか）軌跡を表したものです。


横軸をIterationsとして持っているデータをサンプリングして増やした時に、どのように訓練データと検証データの精度が推移するかを描いたグラフで評価することが多いようです。

learning curveを描いた時に、訓練データに対する精度だけ高かった場合は過学習を疑います。対処法としては、データ量を増やすかモデルが複雑すぎるのでもっと簡単なものを使うかのいずれかです。（もちろん、正則化を入れるというのもありです）

また、データサイズを増やしても精度があがらなければ（accuracyが0.3らへんでさまようとか）、未学習と思えばよいことになります。この場合は、モデルをもっと複雑なものを使うのが良いことになります。


## 正規方程式
thetaを求める方法として、最急降下法(GradientDescent)を使用するのではなく、
正規方程式（Normal Equation）を使用する方法がある。
最急降下法と正規方程式のメリットとデメリットを計算量と計算速度の観点から述べよ。
- 参考
https://qiita.com/antimon2/items/ac1ebaed75ad58406b94

A⊤A が正則なとき，‖Ax−b‖を最小にする xはただ一つであり。
$$ A^{\top}Ax=A^{\top}b $$
を解くことで最小値が求められます。

これを解くと一発で目的関数が最小となるθを求めることができ、シンプルであるというメリットがあります。

しかし、そもそもこの正規方程式では、特徴量が多くなると計算量が増え遅くなるということがあるようです。（行列の次元の三乗で遅くなる）
つまり計算速度は速いが、特徴量が多くなると計算量が増えて対応しきれないという正規方程式のパターンに対して、
特徴量が多くても、うまく動く最急降下法という違いがあるようです。
