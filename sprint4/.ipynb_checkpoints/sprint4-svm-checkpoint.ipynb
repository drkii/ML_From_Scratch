```python
X_df = train_df.drop(["ID","Outcome",],axis=1)
X_df.head()
```

# 標準化を行う

```python
#カテゴリだけ抜き出しとく
X_df_cate = X_df[["Time","Tumor_size","Lymph_node_status"]]
X_df_cate
```

```python
X_df = X_df.drop(["Time","Tumor_size","Lymph_node_status"],axis=1)
X_df.head()
```

## 標準化を行う

```python
X_std = (X_df - X_df.mean() ) / X_df.std()
```

```python
X_std.describe()
```

## 再度合体

```python
X = pd.concat([X_df_cate, X_std],axis=1)
X
```

```python
W = pd.concat([y_df, X],axis=1)
W
```

# 評価

```python
import sklearn.metrics as skm
```

```python
Y_pred_int = np.where(Y_pred == 'R', 1, 0)
Y_pred_int
```

```python
y_test_array = np.array(y_test)
y_test_int =  np.where(y_test_array == 'R', 1, 0)
```

## accuracy

```python
(Y_pred == y_test).sum() / len(Y_pred)
```

```python
skm.accuracy_score(y_test_int, Y_pred_int)
```

## Recall: 再現率 (感度、真陽性率、Recall)

```python
Recall = ( ((Y_pred=='R') & (y_test=='R')).sum() ) / (y_test=='R').sum()
Recall
```

```python
skm.recall_score(y_test_int, Y_pred_int)
```

## Precision

```python
precision = ( ((Y_pred=='R') & (y_test=='R')).sum() )  / (Y_pred=='R').sum()
precision
```

```python
skm.precision_score(y_test_int, Y_pred_int)
```

## F1

```python
F1_score = (2 * Recall * precision) / (Recall + precision)
F1_score
```

```python
skm.f1_score(y_test_int, Y_pred_int)
```
